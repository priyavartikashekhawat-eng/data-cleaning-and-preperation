```python
import pandas as pd
import numpy as np

# -------------------------------------------------------
# STEP 1: Create a messy raw dataset (simulating real-world data)
# -------------------------------------------------------
data = {
    "customer_id": [101, 102, 103, 104, 105, 105, 106, 107, 108, np.nan],
    "name": ["Amit", "neha", "Rohan", "ROHAN", "Sara", "Sara", "Vikram", None, "Pooja", "Karan"],
    "age": [25, 32, np.nan, 29, 41, 41, -5, 30, 27, 200],
    "city": ["Delhi", "mumbai", "Delhi ", "delhi", "Pune", "Pune", "Chennai", "Chennai", None, "Kolkata"],
    "purchase_amount": [1200, 850, 950, 950, np.nan, 950, 3000, 1500, 700, 1100],
}

df = pd.DataFrame(data)

print("STEP 1: Raw Dataset")
print(df, "\n")

# -------------------------------------------------------
# STEP 2: Inspect dataset structure and columns
# -------------------------------------------------------
print("STEP 2: Dataset Info")
print(df.info(), "\n")

print("Missing values per column:")
print(df.isnull().sum(), "\n")

print("Number of duplicate rows (excluding ID):")
print(df.drop(columns=["customer_id"]).duplicated().sum(), "\n")

# -------------------------------------------------------
# STEP 3: Handle missing values
# -------------------------------------------------------

# Drop rows where customer_id is missing (critical identifier)
df = df.dropna(subset=["customer_id"])

# Fill missing names with "Unknown"
df["name"] = df["name"].fillna("Unknown")

# Fill missing age with median age
df["age"] = df["age"].fillna(df["age"].median())

# Fill missing city with "Unknown"
df["city"] = df["city"].fillna("Unknown")

# Fill missing purchase_amount with mean
df["purchase_amount"] = df["purchase_amount"].fillna(df["purchase_amount"].mean())

print("STEP 3: After Handling Missing Values")
print(df, "\n")

# -------------------------------------------------------
# STEP 4: Fix inconsistencies
# -------------------------------------------------------

# Standardize text columns: trim spaces, fix casing
df["name"] = df["name"].str.strip().str.title()
df["city"] = df["city"].str.strip().str.title()

# Fix invalid age values (negative or unrealistic ages)
df["age"] = df["age"].apply(lambda x: np.nan if (x < 0 or x > 100) else x)
df["age"] = df["age"].fillna(df["age"].median())

print("STEP 4: After Fixing Inconsistencies")
print(df, "\n")

# -------------------------------------------------------
# STEP 5: Remove duplicate records
# -------------------------------------------------------
df = df.drop_duplicates(subset=["customer_id"])

print("STEP 5: After Removing Duplicates")
print(df, "\n")

# -------------------------------------------------------
# STEP 6: Final clean, analysis-ready dataset
# -------------------------------------------------------
df = df.reset_index(drop=True)

print("STEP 6: Final Clean Dataset")
print(df, "\n")

print("Final dataset shape:", df.shape)

# Save cleaned data
df.to_csv("cleaned_customer_data.csv", index=False)
print("\nCleaned data saved to 'cleaned_customer_data.csv'")
```
